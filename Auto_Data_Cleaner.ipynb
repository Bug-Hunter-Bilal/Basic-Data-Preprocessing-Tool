# Basic Data Cleaner

```python
# importing required package
import pandas as pd

# loading the file using path provided by the user
def load_file(File_path):
    return pd.read_csv(File_path)
    
# Removing duplicates from the data set using drop_duplicates method
def remove_duplicates(df):
    return df.drop_duplicates()

# Removing columns contains more than 70% null values 
def remove_null_value_columns(df):
    threshold = len(df)*0.30
    total = len(df)
    seventy_percent = total - threshold

    for c in df.columns:
        if((df[c].isna().sum()) > seventy_percent):
            df = df.drop(c, axis = 1)
    return df

# filling the remaining null values with the mean for numerical columns and a string for the object columns
def fill_null_values(df):
    for c in df.columns:
        if(df[c].dtype == 'object') and ((df[c].isna().sum()) != 0):
            df[c] = df[c].fillna("Unknown")
        elif(df[c].dtype == 'int64') and ((df[c].isna().sum()) != 0):
            df[c] = df[c].fillna(df[c].mean())
        elif(df[c].dtype == 'float64') and ((df[c].isna().sum()) != 0):
            df[c] = df[c].fillna(df[c].mean())
    return df

# removing colunms contain more than 70% zeros 
def remove_zero_value_columns(df):
    threshold = len(df)*0.30
    total = len(df)
    seventy_percent = total - threshold
    for c in df.columns:
        if(((df[c] == 0).sum()) > seventy_percent):
            df = df.drop(c, axis = 1)
    return df

# replacing the remaining zero values with the mean
def fill_zero_values(df):
    for c in df.columns:
        if(df[c].dtype == 'int64') and (((df[c] == 0).sum()) != 0):
            df[c] = df[c].replace(0, df[c].mean())
        elif(df[c].dtype == 'float64') and (((df[c] == 0).sum()) != 0):
            df[c] = df[c].replace(0, df[c].mean())
    return df

# Saving the cleaned file
def save_file(df, Output_path):
    df.to_csv(Output_path, index = False)

# main 
if __name__ == "__main__":
    File_path = input("Please enter the complete .csv file path with / slash.")
    Output_file = input('Enter the file name you want with .csv at end')

    df = load_file(File_path)
    print(f"Raw Data:{df.shape}")
    df = remove_duplicates(df)
    df = remove_null_value_columns(df)
    df = fill_null_values(df)
    df = remove_zero_value_columns(df)
    df = fill_zero_values(df)
    save_file(df,Output_file)
    print(f'Preprocessing Completed.\n Final Shape{df.shape}')
```
